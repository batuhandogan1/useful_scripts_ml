```python
import os
```

```python
src_path = '/home/bdogan/all_files/ekin_files/gun_detection/gun_detection_stage_1_dataset/part_2/'

files = os.listdir(src_path)

images_arr = []
labels_arr = []

separated = False

if 'images' in files and 'labels' in files:
    images = os.listdir(os.path.join(src_path, 'images'))
    labels = os.listdir(os.path.join(src_path, 'labels'))
    separated = True

def is_label_empty(label_path):
    return os.path.getsize(label_path) == 0

def get_image_extension(image_name):
    if os.path.exists(os.path.join(src_path, image_name + '.jpg')):
        return '.jpg'
    elif os.path.exists(os.path.join(src_path, image_name + '.png')):
        return '.png'
    return None

if not separated:
    for file in files:
        if file.endswith('.jpg') or file.endswith('.png'):
            splitted_file = os.path.splitext(file)[0]
            images_arr.append(splitted_file)
        elif file.endswith('.txt'):
            splitted_file = os.path.splitext(file)[0]
            labels_arr.append(splitted_file)
else:
    for image in images:
        if image.endswith('.jpg') or image.endswith('.png'):
            splitted_file = os.path.splitext(image)[0]
            images_arr.append(splitted_file)

    for label in labels:
        if label.endswith('.txt'):
            splitted_file = os.path.splitext(label)[0]
            labels_arr.append(splitted_file)

images_arr.sort()
labels_arr.sort()

for image in images_arr:
    image_extension = get_image_extension(image)
    image_path = os.path.join(src_path, image + image_extension) if image_extension else None
    label_path = os.path.join(src_path, f'{image}.txt')

    if image not in labels_arr:
        print(f'{image} does not have a label')
        if image_path:
            os.remove(image_path)

    elif os.path.exists(label_path) and is_label_empty(label_path):
        print(f'{image} label is empty')
        if image_path:
            os.remove(image_path)
        os.remove(label_path)

for label in labels_arr:
    label_path = os.path.join(src_path, f'{label}.txt')

    if label not in images_arr:
        print(f'{label} does not have an image')
        os.remove(label_path)
```
